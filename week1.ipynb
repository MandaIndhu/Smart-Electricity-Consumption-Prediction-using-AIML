```python
import pandas as pd
import numpy as np
import os
```

```python
folder_path = r"C:\Indhu\energy"

dataframes = []
for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        print(f"\nLoaded: {file}, Rows: {df.shape[0]}, Columns: {df.shape[1]}")
        dataframes.append(df)

# Combine all CSVs into one DataFrame (if they have same structure)
combined_df = pd.concat(dataframes, ignore_index=True)
print("\n✅ Combined data shape:", combined_df.shape)
```

```python
combined_df.info()
combined_df.head()
```

```python
combined_df.drop_duplicates(inplace=True)
```

```python
combined_df = combined_df.loc[:, ~combined_df.columns.str.contains('^Unnamed')]
```

```python
# See how many missing values each column has
print(combined_df.isnull().sum())

# Fill missing numeric values with mean
combined_df = combined_df.fillna(combined_df.mean(numeric_only=True))

# OR drop rows with too many missing values
# combined_df = combined_df.dropna()
```
